# Clean the data

Drop duplicates if they exist so they only appear once:

```python
book_ratings_df.drop_duplicates(inplace=True)
books_df.drop_duplicates(inplace=True)
users_df.drop_duplicates(inplace=True)
```

Drop N/A rows from the most important columns:

From book_ratings_df: User-ID, ISBN

From books_df: ISBN,	Book-Title,	Book-Author

From users_df: User-ID

```python
book_ratings_df = book_ratings_df[book_ratings_df['User-ID'].notna()]
book_ratings_df = book_ratings_df[book_ratings_df['ISBN'].notna()]

books_df = books_df[books_df['Book-Title'].notna()]
books_df = books_df[books_df['ISBN'].notna()]
books_df = books_df[books_df['Book-Author'].notna()]

users_df = users_df[users_df['User-ID'].notna()]
```

```python
import re

def clean_text_alphanumeric(text):
    pattern = re.compile('[^a-zA-Z0-9]+')
    return pattern.sub('', text)

def clean_text_alphanumeric_punctuation(text):
    pattern = re.compile('[^a-zA-Z0-9]+ .,:!?/')
    return pattern.sub('', text)

# Apply the function to the dataframe column I want to clean
book_ratings_df['ISBN'] = book_ratings_df['ISBN'].apply(lambda x: clean_text_alphanumeric(x))
books_df['ISBN'] = books_df['ISBN'].apply(lambda x: clean_text_alphanumeric(x))
books_df['Book-Title'] = books_df['Book-Title'].apply(lambda x: clean_text_alphanumeric_punctuation(x))
```

# The most popular books with explicit rating

What makes a book popular? The best way I can answer this is by saying that a book is popular if people enjoy reading it.

Mai intai cele mai populare per general si apoi pe categorii pe varsta: sub 18, 18-25, 25-40, 40+ si locatie

First, we get the average of the ratings per book

```python
average_ratings_per_book = book_ratings_df.groupby(['ISBN'], as_index=False)['Book-Rating'].mean()
```

We want to keep only the most popular books so they should have a average rating above 7.5

```python
book_ratings_above_7_df = average_ratings_per_book.drop(average_ratings_per_book[average_ratings_per_book['Book-Rating'] < 7.5].index)
```

We sort the ratings in descending order so we can see the most popular books

```python
book_ratings_above_7_df.sort_values(['Book-Rating'], ascending=False, inplace=True)
book_ratings_above_7_df
```

In order to get the information about the popular books we found we need to join it with the datagrame that contains informations about all the books

```python
popular_books = pd.merge(book_ratings_above_7_df, books_df, on='ISBN')
popular_books.head(10)
```

```python
# function to get short description from Wikipedia API
def get_description(author):
    url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{author}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if "extract" in data:
            return data['extract']
    return ""
```

For the most popular 1000 authors I will add a short description
